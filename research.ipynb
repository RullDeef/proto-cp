```python
import pandas as pd
import numpy as np
from hvplot import pandas

df1 = pd.read_csv('stat1.csv')
df2 = pd.read_csv('stat2.csv')
df3 = pd.read_csv('stat3.csv')

def add_latency(df):
  df['latency'] = df['recv_time'] - df['send_time']
  #lat = pd.DataFrame(columns=['latency'], data=15.68 * np.ones((len(df1), 1)))
  #df['latency'] = lat['latency']

add_latency(df1)
add_latency(df2)
add_latency(df3)

df1.hvplot(y='latency', ylabel='latency, ms', by='type')
```
